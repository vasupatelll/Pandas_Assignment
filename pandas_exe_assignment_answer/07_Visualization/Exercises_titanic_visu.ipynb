### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv", sep=",")
titanic
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index(titanic["PassengerId"], inplace= True)
titanic
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
male = (titanic["Sex"]=="male").sum()
female=(titanic["Sex"]=="female").sum()
sex=[male,female]
plt.pie(sex,labels=["male","female"],shadow=False, colors=["blue","yellow"], autopct = '%1.1f%%',startangle=90,explode=[0.10,0])
plt.axis("equal")
plt.title("presenting the Male/Female proportion")
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.lmplot(x="Age", y="Fare", data=titanic, hue="Sex")
plt.title("scatterplot with the Fare payed and the Age, differ the plot color by gender")
```

### Step 7. How many people survived?

```python
titanic["Survived"].sum()
```

### Step 8. Create a histogram with the Fare payed

```python
sort = titanic["Fare"].sort_values(ascending=False)
bin_val = np.arange(0,600,10)
plt.hist(sort,bin_val)
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.title("histogram with the Fare payed")
```
